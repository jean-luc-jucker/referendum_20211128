```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Death Rate and Referendum Results
---

## 1.1 Referendum Results

```python
vote = pd.read_excel('je-f-17.03.03.dr.6500.k.xlsx', sheet_name='Cantons', skiprows=8, skipfooter=6, header=None)
vote.rename(columns={1: 'Canton', 10: 'Oui'}, inplace=True)
keep = ['Canton', 'Oui']
vote=vote[keep]
vote.head()
```

```python
# Cantons need to be relabelled for consistency with death rate dataset
def relabel(x):
    if x == 'Zürich':
        label = 'ZH'
    elif x == 'Bern / Berne':
        label = 'BE'        
    elif x == 'Luzern':
        label = 'LU'        
    elif x == 'Uri':
        label = 'UR'        
    elif x == 'Schwyz':
        label = 'SZ'
    elif x == 'Obwalden':
        label = 'OW'
    elif x == 'Nidwalden':
        label = 'NW'
    elif x == 'Glarus':
        label = 'GL'
    elif x == 'Zug':
        label = 'ZG'
    elif x == 'Fribourg / Freiburg':
        label = 'FR'
    elif x == 'Solothurn':
        label = 'SO'
    elif x == 'Basel-Stadt':
        label = 'BS'
    elif x == 'Basel-Landschaft':
        label = 'BL'
    elif x == 'Schaffhausen':
        label = 'SH'
    elif x == 'Appenzell Ausserrhoden':
        label = 'AR'
    elif x == 'Appenzell Innerrhoden':
        label = 'AI'
    elif x == 'St. Gallen':
        label = 'SG'
    elif x == 'Graubünden / Grigioni / Grischun':
        label = 'GR'
    elif x == 'Aargau':
        label = 'AG'
    elif x == 'Thurgau':
        label = 'TG'
    elif x == 'Ticino':
        label = 'TI'
    elif x == 'Vaud':
        label = 'VD'
    elif x == 'Valais / Wallis':
        label = 'VS'
    elif x == 'Neuchâtel':
        label = 'NE'
    elif x == 'Genève':
        label = 'GE'
    elif x == 'Jura':
        label = 'JU'        
    return label
vote['Canton'] = vote['Canton'].apply(relabel)
```

```python
#Visualise
import matplotlib.ticker as mtick
from  matplotlib.ticker import PercentFormatter
sns.set(rc={"figure.dpi":150, 'figure.figsize':(11.7,8.27)}, style='whitegrid', font_scale=1.2)
ax=sns.barplot(data=vote.sort_values(by='Oui', ascending=False), x='Oui', y='Canton', color='gold')
ax.xaxis.set_major_formatter(mtick.PercentFormatter(decimals=0))
ax.axvline(50, color='grey', ls='--')
ch_avg_r = 62.0141101673514
ax.axvline(ch_avg_r, color='blue')
plt.title('Covid Referendum Results by Canton\n', size=18, loc='left')
plt.text(0+0.2, -1, "Blue line: Whole of Switzerland", horizontalalignment='left', size=10, color='blue')
plt.xlabel('Percentage of Yes')
plt.ylabel('Canton')
plt.show()
```

## 1.2 Death rate (since February 2021)

```python
# Note stats exist since February 2021, so this covers whole known period
death = pd.read_csv('COVID19Death_geoRegion.csv')
death.rename(columns={'datum': 'Date', 'geoRegion': 'Canton', 'pop': 'Population', 'inzsumTotal': 'Rate', 'sumTotal': 'Deaths'}, inplace=True)
keep = ['Date', 'Canton', 'Population', 'Deaths', 'Rate']
death=death[keep]
death.head()
```

```python
# Filter out unecessary rows
death = death.loc[(death['Date']=='2021-11-25') & (death['Canton']!='CH') & (death['Canton']!='CHFL') & (death['Canton']!='FL'),
            :]
print(death.shape)
death.head()
```

```python
# Visualise
sns.set(rc={"figure.dpi":150, 'figure.figsize':(11.7,8.27)}, style='whitegrid', font_scale=1.2)
ax=sns.barplot(data=death.sort_values(by='Rate', ascending=False), x='Rate', y='Canton', color='gold')
#ax.axvline(50, color='red')
ch_avg_d = 127.26
ax.axvline(ch_avg_d, color='blue')
plt.title('Covid Deaths per 100,000 People by Canton\n', size=18, loc='left')
plt.text(0+0.2, -1, "Blue line: Whole of Switzerland", horizontalalignment='left', size=10, color='blue')
plt.xlabel('Deaths per 100,000 people\n\n(since February 2020)')
plt.ylabel('Canton')
plt.show()
```

## 1.3 Death Rate and Referendum Results

```python
# We first need to merge data

# Print shapes
print(vote.shape)
print(death.shape)

# Select, sort
vote=vote.sort_values(by='Canton').reset_index(drop=True)
death=death[['Canton', 'Rate']].sort_values(by='Canton').reset_index(drop=True)

# Check Canton is the same
print(vote.Canton.equals(death.Canton))

# Merge
df=pd.merge(vote, death, on='Canton')
df.head()
```

```python
# We add Switzerland as a whole stats
new_line =  {'Canton': 'CH', 'Oui': ch_avg_r, 'Rate': ch_avg_d}
df=df.append(new_line, ignore_index=True)
df.tail()
```

```python
# Quickly see correlation
df.corr()
```

```python
# Visualise
sns.set(rc={"figure.dpi":150, 'figure.figsize':(11.7,8.27)}, style='whitegrid')
ax = sns.regplot(x=df.Rate, y=df.Oui, ci=None, color='gold')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(decimals=0))
plt.title('Covid Referendum Results vs. Death Rate\n', size=18)
#plt.text(153+0.2, 72.5, "Blue point: Whole of Switzerland", horizontalalignment='center', size=10, color='blue')
plt.ylabel('Percentage of Yes\n')
plt.xlabel('\nDeaths per 100,000 people\n\n(since February 2020)')

#Add CH
plt.scatter(df.loc[df['Canton']=='CH', 'Rate'], df.loc[df['Canton']=='CH', 'Oui'], color='blue')
plt.xlim([60, 263])

#Add cut off
#ax.axhline(50, color='grey', ls='--')

for line in range(0,df.shape[0]):
     ax.text(df.Rate[line]+1.5, df.Oui[line], 
     df.Canton[line], horizontalalignment='left', 
     size='medium', color='black')
plt.show()
```

# 2. Vaccination rate vs votation result

```python
# Add Switerland to vote object
new_line =  {'Canton': 'CH', 'Oui': ch_avg_r}
vote=vote.append(new_line, ignore_index=True).sort_values(by='Canton').reset_index(drop=True)
vote.head()
```

```python
vaccinated = pd.read_csv('COVID19VaccPersons_v2.csv')

vaccinated = vaccinated.loc[(vaccinated['geoRegion']!='CHFL') &
               (vaccinated['geoRegion']!='FL') &
               (vaccinated['geoRegion']!='neighboring_chfl') &
               (vaccinated['geoRegion']!='unknown') &
               (vaccinated['geoRegion']!='all') &
               (vaccinated['date']=='2021-11-24') &
               (vaccinated['type']=='COVID19FullyVaccPersons') &
               (vaccinated['age_group']=='total_population')
               , ('geoRegion', 'per100PersonsTotal')].rename(columns={'geoRegion': 'Canton', 'per100PersonsTotal': 'Vaccination'}).reset_index(drop=True)
vaccinated.head()
```

```python
# Check arrays are same before merging
vote.Canton.equals(vaccinated.Canton)
```

```python
# Merge
df=pd.merge(vote, vaccinated, on='Canton')
df.head()
```

```python
df.corr()
```

```python
sns.set(rc={"figure.dpi":150, 'figure.figsize':(11.7,8.27)}, style='whitegrid')
ax = sns.regplot(x=df.Vaccination, y=df.Oui, ci=None, color='gold')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(decimals=0))
ax.xaxis.set_major_formatter(mtick.PercentFormatter(decimals=0))
plt.title('Covid Law Referendum Results vs. Vaccination Rate\n', size=18)
#plt.text(153+0.2, 72.5, "Blue point: Whole of Switzerland", horizontalalignment='center', size=10, color='blue')
plt.ylabel('Percent in Favour of Covid Law\n')
plt.xlabel('\nPercent Fully Vaccinated')

#Add CH
plt.scatter(df.loc[df['Canton']=='CH', 'Vaccination'], df.loc[df['Canton']=='CH', 'Oui'], color='blue')
plt.xlim([53, 71])

#Add cut off
#ax.axhline(50, color='grey', ls='--')

for line in range(0,df.shape[0]):
     ax.text(df.Vaccination[line]+0.2, df.Oui[line], 
     df.Canton[line], horizontalalignment='left', 
     size='medium', color='black')
plt.show()
```
